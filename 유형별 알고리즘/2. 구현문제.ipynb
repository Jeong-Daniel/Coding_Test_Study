### 1. 상하좌우

가장 왼쪽 위 좌표는 (1,1) 가장 오른쪽 좌표는 (N,N)에 해당한다.  
L 왼쪽으로 한칸 이동
R 오른쪽을 한칸 이동  
U 위로 한칸 이동  
D 아래로 한칸 이동  

5  
R R R U D D
->
3 4

```python
n = int(input())
x, y = 1, 1
plans = input().split()

for i in range(len(plans)):
    if plans[i]=='R':
        if x!=n:
            x+=1
    elif plans[i]=='L':
        if x!=1:
            x-=1
    elif plans[i]=='U':
        if y!=1:
            y-=1
    else:
        if y!=n:
            y+=1
    print(y,x)
```

일단 자잘한 것은 다 빼두었다. 예를 들면 소문자가 들어올때 문자 처리라던가
거기에다가 행동을 하는데 있어서 각 방향당 한가지의 제약조건밖에 존재하지 않는다. 예를 들면 x가 1일때는 왼쪽으로 갈 수 없고 n일때는 오른쪽으로 갈 수 없고 해당 조건만 만족하면 카운트 하는 식으로 계산했다. 이런 맵 문제를 생각하면 자연스럽게 2차원배열부터 떠올라서 만들고 볼려고 하는데 이런 간단한 문제는 필요가 없겠다는 생각이 들었다. 시간복잡도는 O(N)이라 한번연산당 한번만 반복하면 되니 어려울 것도 없다. 답지를 보자

보통 이런 문제는 개체를 차례대로 이동시킨다는 점에서 시뮬레이션 유형으로 분류되며 구현이 중요한 대표적인 문제 유형이다. 시뮬레이션, 구현, 완전 탐색 유형등으로 불린다. 또한 코테나 알고리즘 대회에서 가장 난이다고 낮은 1~2번 문제는 대부분 그리디 알고리즘이나 구현 문제이다.

```python
#N을 입력받기
n = int(input())
x,y = 1,1
nx, ny = 0, 0
plans = input().split()

#L,R,U,D에 따른 이동 방향
dx = [0,0,-1,1]
dy = [-1,1,0,0]
move_types = ['L','R','U','D']

#이동 계획을 하나씩 확인
for plan in plans:
    #이동 후 좌표구하기
    for i in range(len(move_types)):
        if plan == move_types[i]:
            nx = x+dx[i]
            ny = y+dy[i]
        #공간을 벗어나는 경우 무시
        if nx < 1 or ny < 1 or nx > n or ny > n:
            continue
        #이동 수행
        x, y = nx, ny

print(x,y)
```

개인적으로는 그냥 내 알고리즘이 더 직관적인거 같다. 그래도 한번 보자.
일단 이중 for문을 도는데 첫번째 for문은 plans의 이동이고 두번째 for문은 LRUD를 리스트형태로 구성해놓고 그에 맞추어서 값을 대입을 하는데 대입한 값이 지도를 벗어나게 되면 반영하지 않고 넘어가는 식으로 구현하였다.

1이 하나라도 포함되어 있을때 세어야하는 시각이다.

```python
h = int(input())
count = 0

for i in range(h+1):
    for j in range(60):
        for k in range(60):
            if '3' in str(i)+str(j)+str(k):
                count +=1
print(count)
```

### 4. 게임 개발

맵은 NxM 크기의 직사각형으로 각 칸은 육지 또는 바다이다 케릭터는 동서남북중 한 곳을 바라보고 있다. 맵의 각 칸은 (A,B)로 나타낼 수 있고, A는 북쪽으로부터 떨어진 칸의 개수, B는 서쪽으로부터 떨어진 칸의 개수이다. 캐릭터는 상하좌우로 움직일 수 있고, 바다로 되어 있는 공간에는 갈 수 없다. 메뉴얼은 다음과 같다.

1. 현재 위치에서 현재 방향을 기준으로 왼쪽 방향(반시계방향으로 90도 회전한 방향) 부터 차례대로 갈곳을 정한다.
2. 캐릭터의 바로 왼쪽 방향에 아직 가보지 않은 칸이 존재한다면 왼쪽방향으로 회전한 다음 왼쪽으로 한칸을 전진한다. 왼족 방향에 가보지 않은 칸이 없다면 왼쪽방향을 회전만 수행하고 1단계로 돌아간다.
3. 만약 네 방향이 모두 이미 가본 칸이거나 바다로 되어 있는 칸인 경우에는, 바라보는 방향을 유지한 채로 한 칸을 뒤로 가고 1단계로 돌아간다. 단 이때 뒤쪽 방향이 바다인 칸이라 뒤로 갈 수 없는 경우에는 움직임을 멈춘다.

위 과정을 반복적으로 수행하면서 캐릭터의 움직임에 이상이 있는지 테스트한다. 매뉴얼에 따라 케릭터를 이동 시킨 다음 캐릭터가 방만한 칸의 수를 출력하는 프로그램을 만드시오  

* 첫째 줄에 맵의 세로 크기 N과 가로 크기 M을 공백으로 구분하여 입력한다. (3<=N M<=50)
* 둘째 줄에 게임 케릭터가 있는 칸의 좌표(A,B)와 바라보는 방향d가 각각 서로 공백으로 구분하여 주어진다. 방향 d의 값으로는 다음과 같이 4가지가 존재한다.
0 북쪽, 1 동쪽, 2 남쪽, 3 서쪽  
* 세째 줄부터 맵이 육지인지 바다인지에 대한 정보가 주어진다. N개의 줄에 맵의 상태가 북쪽부터 남쪽 순서대로, 각 줄의 데이터는 서쪽부터 동쪽 순서대로 주어진다. 맵의 외곽은 항상 바다로 되어 있다. 0:육지, 1:바다

입력예시

1 0 0 1
